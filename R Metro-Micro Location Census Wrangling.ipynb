```python
full_multiracial_df = computed_df.iloc[:,10:].copy()
```

```python
full_multiracial_df[['WFRAC', 'BFRAC', 'AFRAC', 'OFRAC', 'AIANFRAC', 'NHPIFRAC']] = 0
```

full_multiracial_df has 68 columns including totals i.e. "Population of two races:"

```python
# third try, add fracs at end

for column in full_multiracial_df.columns:
    # starting count at 1 because 1 ; splits 2 races
    #cat_total = multiracial_df.loc[index][column]
    if ';' in column:
        fraction = 0
        numchar = 1
        count = column.count(';')
        numchar += count
        fraction += 1/numchar
        print(f'column: {column}')
        print(f'number of races: {numchar}')
        print(f'fraction: {fraction}')
        wfracs = 0
        bfracs = 0
        afracs = 0
        aianfracs = 0
        ofracs = 0
        nhpifracs = 0
        if column.count('White') != 0:
            wfrac = fraction * full_multiracial_df[column]
            wfracs += wfrac
        if column.count('Black or African American') != 0:
            bfrac = fraction * full_multiracial_df[column]
            bfracs += bfrac
        if column.count('American Indian and Alaska Native') != 0:
            aianfrac = fraction * full_multiracial_df[column]
            aianfracs += aianfrac
        if column.count('Asian') != 0:
            afrac = fraction * full_multiracial_df[column]
            afracs += afrac
        if column.count('Some Other Race') != 0:
            ofrac = fraction * full_multiracial_df[column]
            ofracs += ofrac
        if column.count('Native Hawaiian and Other Pacific Islander') != 0:
            nhpifrac = fraction * full_multiracial_df[column]
            nhpifracs += nhpifrac
        full_multiracial_df['WFRAC'] += wfracs
        full_multiracial_df['AFRAC'] += afracs
        full_multiracial_df['OFRAC'] += ofracs
        full_multiracial_df['NHPIFRAC'] += nhpifracs
        full_multiracial_df['AIANFRAC'] += aianfracs
        full_multiracial_df['BFRAC'] += bfracs
    
```

```python
# check the values of a row add up to "Population of two races:" by sum with axis=1
```

```python
# sum components into a total to check it adds to "Population of two or more races:"
full_multiracial_df['FRACTotal'] = full_multiracial_df[['WFRAC', 'BFRAC', 'AFRAC', 'OFRAC', 'AIANFRAC', 'NHPIFRAC']].sum(axis=1).round(2)
```

```python
full_multiracial_df[['Population of two or more races:', 'FRACTotal']]
```

```python
full_multiracial_df['diff_test'] = full_multiracial_df['Population of two or more races:'] - full_multiracial_df['FRACTotal']

(full_multiracial_df['diff_test'] > 0).sum()
```

WFRAC, BFRAC, AFRAC, OFRAC, AIANFRAC, NHPIFRAC are all computed to be added to one-race columns

```python
full_multiracial_df[['WFRAC', 'BFRAC', 'AFRAC', 'OFRAC', 'AIANFRAC', 'NHPIFRAC']] = full_multiracial_df[['WFRAC', 'BFRAC', 'AFRAC', 'OFRAC', 'AIANFRAC', 'NHPIFRAC']].round(3)
```

```python
multirace_mapped_fraction_df = full_multiracial_df[['Population of two or more races:', 'WFRAC', 'BFRAC', 'AFRAC', 'OFRAC', 'AIANFRAC', 'NHPIFRAC', 'FRACTotal']].copy()
```

```python
multirace_mapped_fraction_df.rename(columns={'WFRAC':'W', 'BFRAC':'B', 'OFRAC': 'O', 'AFRAC':'A', 'AIANFRAC':'AIAN', 'NHPIFRAC':'NHPI'}, inplace=True)
```

```python
multirace_mapped_fraction_df.info()
```

```python
completed_df = computed_df.iloc[:,:11].copy().astype(float)
```

```python
completed_df['W'] = multirace_mapped_fraction_df['W'].astype(float).round(3)
completed_df['B'] = multirace_mapped_fraction_df['B'].astype(float).round(3)
completed_df['A'] = multirace_mapped_fraction_df['A'].astype(float).round(3)
completed_df['O'] = multirace_mapped_fraction_df['O']
completed_df['AIAN'] = multirace_mapped_fraction_df['AIAN']
completed_df['NHPI'] = multirace_mapped_fraction_df['NHPI']
```

```python
completed_df['All W'] = completed_df[['White alone', 'W']].sum(axis=1)
completed_df['All B'] = completed_df[['Black or African American alone', 'B']].sum(axis=1)
completed_df['All A'] = completed_df[['Asian alone', 'A']].sum(axis=1)
completed_df['All O'] = completed_df[['Some Other Race alone', 'O']].sum(axis=1)
completed_df['All AIAN'] = completed_df[['American Indian and Alaska Native alone', 'AIAN']].sum(axis=1)
completed_df['All NHPI'] = completed_df[['Native Hawaiian and Other Pacific Islander alone', 'NHPI']].sum(axis=1)
```

```python
for index in completed_df.index:
    for index1 in multirace_mapped_fraction_df.index:
        if index == index1:
            completed_df['All W'] = completed_df['White alone'] + multirace_mapped_fraction_df['W']
```
