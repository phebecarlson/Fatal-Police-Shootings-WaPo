full_multiracial_df has 68 columns including totals i.e. "Population of two races:"

```python
for column in multiracial_df.columns:
    # starting count at 1 because 1 ; splits 2 races
    #cat_total = multiracial_df.loc[index][column]
    if ';' in column:
        fraction = 0
        numchar = 1
        count = column.count(';')
        numchar += count
        fraction += 1/numchar
        print(f'column: {column}')
        print(f'number of races: {numchar}')
        print(f'fraction: {fraction}')
    if column.count('White') != 0:
        wfrac = fraction * multiracial_df[column]
        print(wfrac)
        multiracial_df['WFRAC'] += wfrac
        #print(column.count('White'))
    if column.count('Black or African American') != 0:
        bfrac = fraction * multiracial_df[column]
        print(bfrac)
        multiracial_df['BFRAC'] += bfrac
    if column.count('American Indian and Alaska Native') != 0:
        aianfrac = fraction * multiracial_df[column]
        print(aianfrac)
        multiracial_df['AIANFRAC'] += aianfrac
    if column.count('Asian') != 0:
        afrac = fraction * multiracial_df[column]
        print(afrac)
        multiracial_df['AFRAC'] += afrac
        #print(column.count('Asian'))
    if column.count('Some Other Race') != 0:
        ofrac = fraction * multiracial_df[column]
        print(ofrac)
        multiracial_df['OFRAC'] += ofrac
    if column.count('Native Hawaiian and Other Pacific Islander') != 0:
        nhpifrac = fraction * multiracial_df[column]
        print(nhpifrac)
        multiracial_df['NHPIFRAC'] += nhpifrac
```
