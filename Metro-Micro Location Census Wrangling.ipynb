```python
# responses: one race alone and "2 or more races:"
one_race_df = computed_df.iloc[:,:11].copy()
```

```python
full_multiracial_df = computed_df.iloc[:,10:].copy()
```

```python
full_multiracial_df[['WFRAC', 'BFRAC', 'AFRAC', 'OFRAC', 'AIANFRAC', 'NHPIFRAC']] = 0
```

full_multiracial_df has 68 columns including totals i.e. "Population of two races:"

```python
for column in full_multiracial_df.columns:
    # starting count at 1 because 1 ; splits 2 races
    #cat_total = multiracial_df.loc[index][column]
    if ';' in column:
        fraction = 0
        numchar = 1
        count = column.count(';')
        numchar += count
        fraction += 1/numchar
        print(f'column: {column}')
        print(f'number of races: {numchar}')
        print(f'fraction: {fraction}')
        wfracs = 0
        bfracs = 0
        afracs = 0
        aianfracs = 0
        ofracs = 0
        nhpifracs = 0
        if column.count('White') != 0:
            wfrac = fraction * full_multiracial_df[column]
            wfracs += wfrac
        if column.count('Black or African American') != 0:
            bfrac = fraction * full_multiracial_df[column]
            bfracs += bfrac
        if column.count('American Indian and Alaska Native') != 0:
            aianfrac = fraction * full_multiracial_df[column]
            aianfracs += aianfrac
        if column.count('Asian') != 0:
            afrac = fraction * full_multiracial_df[column]
            afracs += afrac
        if column.count('Some Other Race') != 0:
            ofrac = fraction * full_multiracial_df[column]
            ofracs += ofrac
        if column.count('Native Hawaiian and Other Pacific Islander') != 0:
            nhpifrac = fraction * full_multiracial_df[column]
            nhpifracs += nhpifrac
        full_multiracial_df['WFRAC'] += wfracs
        full_multiracial_df['AFRAC'] += afracs
        full_multiracial_df['OFRAC'] += ofracs
        full_multiracial_df['NHPIFRAC'] += nhpifracs
        full_multiracial_df['AIANFRAC'] += aianfracs
        full_multiracial_df['BFRAC'] += bfracs
    
```

```python
# check the values of a row add up to "Population of two races:" by sum with axis=1
```

```python
# sum components into a total to check it adds to "Population of two or more races:"
full_multiracial_df['FRACTotal'] = full_multiracial_df[['WFRAC', 'BFRAC', 'AFRAC', 'OFRAC', 'AIANFRAC', 'NHPIFRAC']].sum(axis=1).round(2)
```

```python
full_multiracial_df[['Population of two or more races:', 'FRACTotal']]
```

```python
full_multiracial_df['diff_test'] = full_multiracial_df['Population of two or more races:'] - full_multiracial_df['FRACTotal']

(full_multiracial_df['diff_test'] > 0).sum()
```

All of the rows where FRACTotal isn't exact are off by exactly 1. This is probably a rounding or math error somewhere when I changed from int to float. 

The differences turned out to be fractions of a hundredth. Needed to round float data.

```python
# check the math to see if each column calculated properly and can be added back in
for index in full_multiracial_df.index:
    row = full_multiracial_df.loc[index]
    if row['Population of two or more races:'] == row['FRACTotal']:
        print('y')
    else: print(index)
```
