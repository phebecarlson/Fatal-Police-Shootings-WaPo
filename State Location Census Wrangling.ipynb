full_multiracial_sdf has 69 columns including totals i.e. "Population of two races:"

Fraction out State Proportions

```python
for column in full_multiracial_sdf.columns:
    # starting count at 1 because 1 ; splits 2 races
    if ';' in column:
        fraction = 0
        numchar = 1
        count = column.count(';')
        numchar += count
        fraction += 1/numchar
        print(f'column: {column}')
        print(f'number of races: {numchar}')
        print(f'fraction: {fraction}')
        wfracs = 0
        bfracs = 0
        afracs = 0
        aianfracs = 0
        ofracs = 0
        nhpifracs = 0
        if column.count('White') != 0:
            wfrac = fraction * full_multiracial_sdf[column]
            wfracs += wfrac
        if column.count('Black or African American') != 0:
            bfrac = fraction * full_multiracial_sdf[column]
            bfracs += bfrac
        if column.count('American Indian and Alaska Native') != 0:
            aianfrac = fraction * full_multiracial_sdf[column]
            aianfracs += aianfrac
        if column.count('Asian') != 0:
            afrac = fraction * full_multiracial_sdf[column]
            afracs += afrac
        if column.count('Some Other Race') != 0:
            ofrac = fraction * full_multiracial_sdf[column]
            ofracs += ofrac
        if column.count('Native Hawaiian and Other Pacific Islander') != 0:
            nhpifrac = fraction * full_multiracial_sdf[column]
            nhpifracs += nhpifrac
        full_multiracial_sdf['WFRAC'] += wfracs
        full_multiracial_sdf['AFRAC'] += afracs
        full_multiracial_sdf['OFRAC'] += ofracs
        full_multiracial_sdf['NHPIFRAC'] += nhpifracs
        full_multiracial_sdf['AIANFRAC'] += aianfracs
        full_multiracial_sdf['BFRAC'] += bfracs
    
```
